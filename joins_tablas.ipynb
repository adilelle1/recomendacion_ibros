# concateno 

```python
import numpy as np
```

```python
book_data.columns
```

```python
book_data['isbn'] = np.NaN
```

```python
data_meli.columns
```

```python
data_nueva.columns
```

```python
data_nueva.drop(columns=['bookformat', 'img','link', 'rating', 'reviews', 'totalratings'], inplace=True)
data_meli.drop(columns=['subtitle','thumbnail','published_year', 'average_rating', 'ratings_count'], inplace=True)
book_data.drop(columns=['book_edition', 'book_format', 'book_rating', 'book_rating_count', 'book_review_count', 'image_url'], inplace=True)
```

```python
data_nueva.rename(columns={'desc': 'description', 'author':'authors'}, inplace=True)
data_meli.rename(columns={'isbn10': 'isbn','num_pages':'pages','categories':'genre'}, inplace=True)
book_data.rename(columns={'book_authors':'authors', 'book_desc':'description', 'book_isbn':'isbn13', 'book_title':'title','book_pages':'pages','genres':'genre'}, inplace=True)
```

```python
concatenated_df = pd.concat([book_data, data_nueva, data_meli])
concatenated_df
```

```python
concatenated_df.loc[concatenated_df['pages'].str.contains('pages', na=False), 'pages'] = concatenated_df.loc[concatenated_df['pages'].str.contains('pages', na=False), 'pages'].str.extract('(\d+)', expand=False).astype(int)
concatenated_df
```

```python
concatenated_df.title.nunique()
```

```python
concatenated_df.drop_duplicates(subset='title', keep='first', inplace=True)

```

```python
concatenated_df
```

```python
data2.loc[data2['description'].isnull(),'title']= data2.loc[data2['description'].isnull(),'title'].apply(lambda x: x.split('(', 1)[0].strip() if '(' in x else x)
data2.loc[data2['description'].isnull(),'title']
```

```python
merged_df = data2.merge(concatenated_df[['title', 'description','genre', 'pages']], on='title', how='left')
data2['description'] = merged_df['description_y'].fillna(data2['description'])
data2['genre'] = merged_df['genre']
data2['pages'] = merged_df['pages']
```

```python
merged_df = data2.merge(concatenated_df[['isbn', 'description']], on='isbn', how='left')
data2['description'] = merged_df['description_y'].fillna(data2['description'])
data2['genre'] = merged_df['genre']
data2['pages'] = merged_df['pages']
```

```python
merged_df = data2.merge(concatenated_df[['isbn13', 'description']], on='isbn13', how='left')
data2['description'] = merged_df['description_y'].fillna(data2['description'])
data2['genre'] = merged_df['genre']
data2['pages'] = merged_df['pages']
```
