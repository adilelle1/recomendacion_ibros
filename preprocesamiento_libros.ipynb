```python
import pandas as pd

data_nueva = pd.read_csv('data/GoodReads_100k_books.csv')
data_nueva.head()
```

```python
data_nueva.dropna(subset=['desc'], inplace=True)
data_nueva.isnull().sum()
```

```python
data_meli = pd.read_csv('data/books.csv')
data_meli.head()
```

```python
data_meli.dropna(subset=['description'], inplace=True)
data_meli.isnull().sum()
```

```python
data2 = pd.read_csv('data/books2.csv')
data2.head()
```

```python
data2.drop(columns=['image_url','small_image_url'])
```

```python
import numpy as np
data2['description'] = np.nan
data2['description'].isnull().sum()
```

```python
data2.drop_duplicates(subset='isbn13', keep='first', inplace=True)
data2.drop_duplicates(subset='title', keep='first', inplace=True)
data2.drop_duplicates(subset='isbn', keep='first', inplace=True)
```

```python
book_data = pd.read_csv('data/book_data.csv')
book_data.head()
```

```python
book_data.drop_duplicates(subset='book_title', keep='first', inplace=True)
```

# concateno 

```python
book_data['isbn'] = np.NaN
```

```python
data_nueva.drop(columns=['bookformat', 'img','link', 'rating', 'reviews', 'totalratings'], inplace=True)
data_meli.drop(columns=['subtitle','thumbnail','published_year', 'average_rating', 'ratings_count'], inplace=True)
book_data.drop(columns=['book_edition', 'book_format', 'book_rating', 'book_rating_count', 'book_review_count', 'image_url'], inplace=True)
```

```python
data_nueva.rename(columns={'desc': 'description', 'author':'authors'}, inplace=True)
data_meli.rename(columns={'isbn10': 'isbn','num_pages':'pages','categories':'genre'}, inplace=True)
book_data.rename(columns={'book_authors':'authors', 'book_desc':'description', 'book_isbn':'isbn13', 'book_title':'title','book_pages':'pages','genres':'genre'}, inplace=True)
```

```python
concatenated_df = pd.concat([book_data, data_nueva, data_meli])
concatenated_df
```

```python
concatenated_df.loc[concatenated_df['pages'].str.contains('pages', na=False), 'pages'] = concatenated_df.loc[concatenated_df['pages'].str.contains('pages', na=False), 'pages'].str.extract('(\d+)', expand=False).astype(int)
concatenated_df
```

```python
concatenated_df.title.nunique()
```

```python
concatenated_df.drop_duplicates(subset='title', keep='first', inplace=True)

```

```python
concatenated_df['genre']=concatenated_df['genre'].str.replace("|", ",")
```

```python
genre_counts = concatenated_df['genre'].str.split(',').explode().str.strip().value_counts()
genre_counts
```

```python
ordered_genres = genre_counts.index.tolist()

concatenated_df['genre_ordenado'] = concatenated_df['genre'].apply(lambda x: ', '.join(sorted(str(x).split(','), key=lambda g: ordered_genres.index(g.strip())) if isinstance(x, str) else ''))
```

```python
concatenated_df[['genre_ordenado','genre']]
```

```python
genre_columns = concatenated_df["genre_ordenado"].str.split(",", expand=True)
genre_columns = genre_columns.iloc[:, :4]  

concatenated_df = pd.concat([concatenated_df, genre_columns], axis=1)
```

```python
data2.loc[data2['description'].isnull(),'title']= data2.loc[data2['description'].isnull(),'title'].apply(lambda x: x.split('(', 1)[0].strip() if '(' in x else x)
data2.loc[data2['description'].isnull(),'title']
```

```python
concatenated_df.rename(columns = {0:'genero_1'}, inplace = True)
concatenated_df.rename(columns = {1:'genero_2'}, inplace = True)
concatenated_df.rename(columns = {2:'genero_3'}, inplace = True)
concatenated_df.rename(columns = {3:'genero_4'}, inplace = True)
```

```python
merged_df = data2.merge(concatenated_df[['title', 'description','pages', 'genre', 'genre_ordenado','genero_1','genero_2','genero_3','genero_4']], on='title', how='left')
merged_df.head()
```

```python
merged_df.isnull().sum()
```

```python
concatenated_df_sin_merge = concatenated_df[~concatenated_df['title'].isin(merged_df['title'])].copy()
concatenated_df_sin_merge
```

```python
concatenated_df_sin_merge.isnull().sum()
```

```python
concatenated_df_sin_merge.drop(columns=['isbn','isbn13'], inplace=True)
concatenated_df_sin_merge.dropna(subset=['genero_4'], inplace=True)
concatenated_df_sin_merge.dropna(subset=['description'], inplace=True)
concatenated_df_sin_merge.dropna(subset=['pages'], inplace=True)
concatenated_df_sin_merge.isnull().sum()
```

```python
concatenated_df_sin_merge.shape
```

```python
# guardo los libros de los otros datasets concatenados, ya procesado pero no tienen ratings
concatenated_df_sin_merge.to_csv('data/df_libros_sin_rating.csv')
```

# Ultimos ajustes sobre el df final

```python
merged_df.drop(columns=['best_book_id', 'work_id', 'image_url','small_image_url', 'description_x'], inplace=True)
```

```python
merged_df.isnull().sum()
```

```python
merged_df.dropna(subset= ['genero_4', 'description_y','pages','isbn'],inplace=True)
merged_df.isnull().sum()
```

```python
from langdetect import detect

# Define una función para detectar el idioma de un texto
def detect_language(text):
    try:
        language = detect(text)
        return language
    except Exception as e:
        print(f"Error al detectar el idioma: {e}")
        return "No detectado"

# Aplica la detección de idioma a la columna 'descripcion'
merged_df["description_y"] = merged_df["description_y"].astype(str)
merged_df.loc[merged_df['language_code'].isnull(),'language_code'] = merged_df.loc[merged_df['language_code'].isnull(),'description_y'].apply(detect_language)
```

```python
merged_df['language_code'] = merged_df['language_code'].apply(lambda x: 'eng' if 'en' in x else x)
```

```python
merged_df.loc[merged_df['original_title'].isnull(), 'original_title'] = merged_df.loc[merged_df['original_title'].isnull(), 'title']
```

```python
publication_year_to_fill = merged_df.loc[merged_df['original_publication_year'].isnull(), 'title']
```

```python
merged_df.rename(columns = {'description_y':'description'}, inplace = True)
```

```python
titles_sin_year = merged_df.loc[merged_df.original_publication_year.isnull(), 'title'].tolist()
titles_sin_year
```

```python
publication_years = [2008, 2003, 2009, 2004, 1861, 1974, 1939, 2007]
```

```python
for title, year in zip(titles_sin_year, publication_years):
    merged_df.loc[merged_df['title'] == title, 'original_publication_year'] = year
```
